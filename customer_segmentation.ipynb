```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering

df = pd.read_csv("/content/Mall_Customers.csv")

x = df[['Annual Income (k$)','Spending Score (1-100)']]

agg_model = AgglomerativeClustering(n_clusters=3)

df['Cluster'] = agg_model.fit_predict(x)

plt.scatter(x['Annual Income (k$)'],x['Spending Score (1-100)'],c=df['Cluster'],cmap='rainbow')
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.title("Agglomerative Clustering Example")
plt.show()

df.to_csv("agglo_clustered data.csv",index=False)
print("Agglomerative clusters dataset saved as 'agglo_clustered_data.csv'")
```
